```python
import datetime
import numpy as np
```

```python
start = datetime.datetime.now()
end = datetime.datetime(2017, 11, 20, start.hour+3, 59, 59)
step = datetime.timedelta(seconds=30)

result = []

while start < end:
    result.append(start.isoformat()+"Z")
    start += step
date_length = len(result)
```
